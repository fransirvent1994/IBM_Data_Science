# Instrucciones de uso de git básicos

## Git es una herramienta de comandos de automatización que usa Linux. Es de código abierto y una gran versatilidad y usos. Es una herramienta fantástica para control de versiones, automatización, etc.

## Para instalar git en windows: https://gitforwindows.org/

## Una vez instalado, con click derecho en el escritorio, abrimos "Git Bash", que abrirá una ventana de comandos. Si escribimos "git --version" debería salir la versión instalada de Git.

### Con esta herramienta ya instalada, podemos mejorar nuestro flujo de trabajo de una manera muy efectiva. Se podrá subir múltiples archivos en plataformas como GitHub con unos pocos comandos. Esto significa que, una vez creado y vinculado un repositorio, un proyecto de múltiples archivos de código, generalmente lo subiriamos uno a uno, sin embargo, con Git podemos subirlos todos de golpe, incluyendo los commit y sus tags.

#### Esto aplica su uso también en otras plataformas, como Visual Studio Code o Pycharm, los cuales tienen su propia terminal, y se puede usar Git.

# Instrucciones de comandos en git para subir en repertorio de GitHub:

## Crearemos un repositorio que vincular a la cuenta de Git.
### Cuando creamos un repositorio, se crean una serie de instrucciones para el usuario para facilitar las cosas.

<img src="InstruccionesGitHub.png">

## Habiendo creado este repositorio, hará falta introducir estos comandos en git.
## Para ello haremos lo siguiente:
1. Creamos una carpeta (preferiblemente con el mismo nombre que el repositorio, para evitar confusiones, o una carpeta que almacene esta misma)
2. En este directorio abriremos Git Bash con click derecho, e introduciremos git init.
3. Una vez introducido git init, se creará dentro del directorio una carpeta .git, la cual permitirá generar instrucciones entre git y github.
4. Para comprobar si existe esa carpeta, en el caso de que no la veais, introducid el comando ls -a en git, que os mostrará la lista de carpetas en ese directorio, pero también muestra las ocultas, y debería salir .git
6. Ya comprobado eso, abriremos con git code un archivo. Pongo ejemplo de comando y explico:
   * git code "File1" Explico: git es el comando principal. Code para abrir un archivo de texto. File1 es el nombre del archivo de texto, puedes elegir el nombre que quieras.
7. Una vez creado este archivo, procderemos a usar los siguientes comandos:
   * git remote add origin https://github.com/nombredecuenta/nombrederepositorio.git Explico: Con este comando se vincula la carpeta de código con el repositorio de GitHub.
   * git add . Explico: Con este comando, todo lo creado en esa carpeta se tendrá en cuenta para las próximas acciones.
   * git commit -m "My first commit" Explico: Con este comando hacemos el commit. Lo anterior descrito con add, será metido dentro de esta acción, procediendo a darle el nombre elegido en el interior de las comillas, que describirá generalmente los cambios realizados.
   * git push -u origin main Explico: Todo lo anterior se subirá al repositorio vinculado.

### Notas:
1. Posiblemente os dé errores en varias ocasiones. No pasa nada. Es un sistema completo, te señala donde está el fallo, es cuestión de fijarse.
2. IMPORTANTE: Por lo general, los repositorios de GitHub se crean con la etiqueta/branch/rama (master). Es preferible cambiarlo a "main", por lo tanto, tras haber vinculado la cuenta con git remote, se usará git branch -M main para convertir la rama de master en main.
